```python
import csv

def count_values(csv_file_path):
    values_counts = {}

    with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        
        
        for row in reader:
            
            for column, value in row.items():
                
                if column == 'description':
                    continue

                
                if column == 'date_added' or column == 'title':
                    values = [value.strip()]  
                else:
                    values = [v.strip() for v in value.split(',')]
                
                if column not in values_counts:
                    values_counts[column] = {}
                
                
                for v in values:
                    values_counts[column][v] = values_counts[column].get(v, 0) + 1

    for column, counts in values_counts.items():
        print(f"Values count for {column}:")
        total_values = len(counts)
        print(f"Total different values for {column} is : {total_values}")
        max_value = max(counts, key=counts.get)
        max_count = counts[max_value]
        # If the max value is empty cuz sometimes we dont have the data for that speciifc column, find the second max value
        if max_value == '':
            second_max_value = max(counts, key=lambda k: (counts[k], k) if k != max_value else (0, ''))
            max_value = second_max_value
            max_count = counts[max_value]

        print(f"Max count for {column}: {max_count} ({max_value})")
        for value, count in counts.items():
            print(f"{value}: {count}")
        print("--------------------------------------------")
        print()
```

```python
count_values('netflix_titles.csv')
```
